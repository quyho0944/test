##INTRODUCTION






Viliuisk encephalomyelitis (VE) is a fatal progressive neurological disorder found only in the Sakha (Iakut/Yakut) population of central Siberia.[1][2] About 15 new cases are reported each year. VE is a very rare disease and little research has been conducted. The causative agents, origin of the disease, and involved candidate genes are currently unknown.

## PATHOPHYSIOLOGY

The disease has three distinguishable phases: The acute form, the progressive form, and the chronic form





*   The acute form is the most rapid and most violent of all the stages. It begins with the characteristic rigidity of the muscles, accompanied by slurred speech, severe headaches, and exaggeration of cold-like symptoms. Patients usually die within weeks of the initial symptoms. Routine post-mortem examinations yield: severe inflammation of the brain lining, clusters of dead cells and tissue, and largely increased amounts of macrophages and lymphocytes
*   The progressive form is the most common case. Patients initially experience acute-like symptoms which are not as severe, and subside within a few weeks. Following the sub-acute phase, the patients experience a few mild symptoms including some behavioral changes, incoordination, and difficulty in speech. Eventually the disease developed fully and those infected were stricken with the characteristic symptoms of rigidity, slurred speech, and deterioration of cognitive functions.[3][4] Ultimately, brain function depreciates rapidly resulting in death.
*     the chronic form claim never to have had an acute attack. These patients endure varying measures of impairment and suffer mental deterioration for the remainder of their lives. Usually they live to be very old and succumb to other diseases.

## Concat files

Combine the files altogether

````python
%%shell

for i in trimmed_*.fastq.gz; do \
    id=$(echo $i | cut -d'_' -f2);
    awk -v id=$id '/^>/ \
        {getline seq; a[count++][$0]=seq} \
        END {label=10^length(count); \
        for (i in a) {\
            for (j in a[i]) { \
                print ">" id "_" label+i "\n" a[i][j]} \
            }
        }' <(zcat $i | \
            awk 'NR%4==1 {sub(/^@/, ">", $0); print; getline; print}'); \
done > total.fasta
````

### Read mapping

````python
%%shell

for i in trimmed_*fastq.gz; do \
    id=$(echo $i | cut -d'_' -f2); \
    echo -e "########\n${i/.lite.1*/}"; \
    bowtie2 -x dedup_total -q $i -S $id.sam -p 4; \
done 2>&1 | tee bowtie2.report
````

### Retrieve coverage

````python
%%shell

for id in $(ls *.sam | cut -d'.' -f1); do \
    samtools view -bS ${id}.sam > ${id}.bam; \
    samtools sort ${id}.bam -@ 8 -o ${id}_sorted.bam; \
    samtools index ${id}_sorted.bam; \
    samtools depth ${id}_sorted.bam > ${id}_coverage.txt; \
done
````

### Retrieve total depth

````python
%%shell

for i in *_coverage.txt; do \
    awk -v FS="\t" ' \
    {sum[$1]+=$3} \
    END {for (i in sum) print i "\t" sum[i]}' $i > ${i%%_coverage.txt}_depth.txt; \
done
````

### Retrieve TPM

````python
%%shell

for i in *_depth.txt; do \
    awk 'fname!=FILENAME {fname=FILENAME; idx++} \
    idx==1 {if ($0~/^>/) {getline seq; gsub(/^>/, "", $0); len[$0]=length(seq)}} \
    idx==2 {FS="\t"; depth[$1]=$2; sum+=$2} \
    END {for (i in depth) print i "\t" depth[i]*1e9/(sum*len[i])}' dedup_total.fasta $i > ${i/depth/tpm}; \
done
````

## Generate count table

````python
%%shell

awk 'BEGIN {FS=OFS="\t"} \
    {a[$1][ARGIND]=$2; filenames[ARGIND]=FILENAME} \
    END {printf "ID%s", OFS; \
        for (x=1; x<=ARGIND; x++) {printf "%s%s", filenames[x], (x<ARGIND ? OFS : ORS)}; \
        for (i in a) {printf "%s%s", i, OFS; \
            for (j=1; j<=ARGIND; j++) { \
                k=(j in a[i] ? a[i][j] : 0); \
                printf "%s%s", k, (j<ARGIND ? OFS : ORS)
                }
            }
        }' *_tpm.txt > count_table.tsv
````

Keep only results with highest percent identity and query coverage

````python
%%shell

awk '\
    function abs(n) {return (n>0 ? n : -n)} \
    BEGIN {FS=OFS="\t"} \
    ($3>max_pident[$1] && abs($(NF-3)-$(NF-2)+1)/$5>=max_qcov[$1]) {max_pident[$1]=$3; max_qcov[$1]=abs($(NF-3)-$(NF-2)+1)/$5; a[$1]=$0} \
    END {for (i in a) print a[i]}' dedup_total.tsv > process_dedup_total.tsv
````

Filter sequences with 95% similarity and 100% query coverage

````python
  %%shell

  awk 'function abs(n) {return (n>0 ? n : -n)} \
      BEGIN {FS=OFS="\t"} \
      {$3>95 && abs($(NF-3)-$(NF-2)+1)/$5>=1.0} {print}' process_dedup_total.tsv > filter_process_dedup_total.tsv
````

````python
import pandas as pd
from skbio.diversity import alpha_diversity, beta_diversity
from skbio.stats.ordination import pcoa
import matplotlib.pyplot as plt

# Read count table into a Pandas DataFrame
count_table_yakuts = pd.read_csv('/content/data_yakult/count_table_yakuts.tsv', sep='\t')
count_table_citizen = pd.read_csv('/content/count_table_k1000.tsv', sep='\t')
# Transpose the DataFrame to have samples as rows and features as columns
# Scikit-bio only accepts this format
transposed_table_yakuts = count_table_yakuts.T
transposed_table_citizen = count_table_citizen.T

# Convert count values to numeric
transposed_table_yakuts = transposed_table_yakuts.apply(pd.to_numeric, errors='coerce')
transposed_table_citizen = transposed_table_citizen.apply(pd.to_numeric, errors='coerce')


# Remove rows with NaN values (if any)
transposed_table_yakuts = transposed_table_yakuts.dropna()
transposed_table_citizen = transposed_table_citizen.dropna()
# Calculate alpha diversity (Shannon entropy in this example)
alpha_div_yakuts = alpha_diversity('shannon', transposed_table_yakuts, ids=transposed_table_yakuts.index)
alpha_div_citizen = alpha_diversity('shannon', transposed_table_citizen, ids=transposed_table_citizen.index)

print("Alpha Diversity:")
print(alpha_div_yakuts)
print(alpha_div_citizen)
````

````python
# Print the calculated alpha diversity
print("Alpha Diversity:")
print(type(alpha_div_yakuts.T))
print("Columns:", alpha_div_yakuts['SRR10261561'])
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a sample DataFrame (replace this with your own data)
data_citizen = {
    'citizen': alpha_div_citizen.tolist(),
}

data_yakuts = {
    'yakuts': alpha_div_yakuts.tolist()
}

df_citizen = pd.DataFrame(data_citizen)
df_yakuts = pd.DataFrame(data_yakuts)
list_data = [df_citizen, df_yakuts]
# Plotting
for df in list_data:
  sns.boxplot(data=df)
  plt.title('Alpha Diversity Boxplot')
  plt.ylabel('Alpha Diversity')
  plt.show()
````

Furthermore, the median alpha diversity value for the Yakuts group surpasses 13.5, exceeding the corresponding median of approximately 12.5 for the Citizen group. This disparity in medians underscores a substantial difference in the central tendency of alpha diversity between the two groups.

The widened IQR in the Yakuts boxplot implies a more even distribution of alpha diversity values compared to the Citizen group. This heightened diversity, coupled with a lack of outliers, suggests a more homogeneous and consistently elevated alpha diversity within the Yakuts population in contrast to the Citizen group.

In summary, the analytical findings indicate a distinct and more evenly distributed alpha diversity within the Yakuts group, surpassing that observed in the Citizen group.
